```python
#importing libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
import matplotlib.pyplot as plt
```

```python
#Reading csv into dataframe
df=pd.read_csv(r"C:\Users\ddnri\OneDrive\creditcard.csv")
#Printing data frame
df
```

```python
#shows class imbalance
df['Class'].value_counts()
```

```python
#Handling imbalanced data, Making dataset balanced
x=df.drop('Class',axis=1)
y=df['Class']
sm=SMOTE()
x_res,y_res=sm.fit_resample(x,y)
print(pd.Series(y_res).value_counts())
```

```python
#checking null values in each column
df.isnull().sum()
```

```python
#Removing Duplicate rows
df.drop_duplicates(inplace=True)
#Printing column names
df.columns
```

```python
#Performing Standardisation 
ss=StandardScaler()
df['Time']=ss.fit_transform(df[['Time']])
df['Amount']=ss.fit_transform(df[['Amount']])
df
```

```python
#Handling Outliers using z-score method
for col in df.columns:
    mu=df[col].mean()
    sig=df[col].std()
    df['z-score']=(df[col]-mu)/sig
    df[(df['z-score']>-3)&(df['z-score']<3)]
df
```

```python
#Handling Outliers using iqr method
for col1 in df.columns:
    q1=df[col1].quantile(0.25)
    q3=df[col1].quantile(0.75)
    iqr=q3-q1
    upp=q3+1.5*iqr
    low=q1-1.5*iqr
    df[(df[col1]>=low)&(df[col1]<=upp)]
df
```

```python
#Ploting a histogram on class column
plt.hist(df['Class'])
plt.title("Histogram on class column")
plt.show()
```
